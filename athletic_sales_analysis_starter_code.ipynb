```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_2020_2021_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", join="inner")
```

```python
# Check if any values are null.
athletic_sales_2020_2021_df
```

```python
# Check the data type of each column
athletic_sales_2020_2021_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_2020_2021_df.invoice_date = pd.to_datetime(athletic_sales_2020_2021_df.invoice_date, format="%m/%d/%y")
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_2020_2021_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
athletic_sales_2020_2021_grouped_df = athletic_sales_2020_2021_df.groupby(["region", "state", "city"]).agg({"units_sold": "sum"}).rename(columns={"units_sold": "Total_Products_Sold"})
# Show the top 5 results.
athletic_sales_2020_2021_grouped_df.sort_values("Total_Products_Sold", ascending=False).head(5)              
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
athletic_sales_2020_2021_pivoted_df = pd.pivot_table(athletic_sales_2020_2021_df, index=["region", "state", "city"], values=["units_sold"], aggfunc=sum)

# Rename the "units_sold" column to "Total_Products_Sold"
athletic_sales_2020_2021_pivoted_df.rename(columns={"units_sold": "Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_pivoted_df.sort_values("Total_Products_Sold", ascending=False).head(5)

```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_2020_2021_total_sales_df = athletic_sales_2020_2021_df.groupby(["region", "state", "city"]).agg({"total_sales": "sum"})

# Rename the "total_sales" column to "Total Sales"
athletic_sales_2020_2021_total_sales_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_total_sales_df.sort_values("Total Sales", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_2020_2021_sales_pivoted_df = pd.pivot_table(athletic_sales_2020_2021_df, index=["region", "state", "city"], values=["total_sales"], aggfunc=sum)

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_2020_2021_sales_pivoted_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_sales_pivoted_df.sort_values("Total Sales", ascending=False).head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
athletic_sales_2020_2021_total_sales_retailer_df = athletic_sales_2020_2021_df.groupby(["retailer", "region", "state", "city"]).agg({"total_sales": "sum"})

# Rename the "total_sales" column to "Total Sales"
athletic_sales_2020_2021_total_sales_retailer_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_total_sales_retailer_df.sort_values("Total Sales", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
athletic_sales_2020_2021_sales_retailer_pivoted_df = pd.pivot_table(athletic_sales_2020_2021_df, index=["retailer", "region", "state", "city"], values=["total_sales"], aggfunc=sum)

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_2020_2021_sales_retailer_pivoted_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_sales_retailer_pivoted_df.sort_values("Total Sales", ascending=False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
athletic_sales_2020_2021_womens_df = athletic_sales_2020_2021_df[athletic_sales_2020_2021_df['product'] == 'Women\'s Apparel']
athletic_sales_2020_2021_womens_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_2020_2021_womens_retailer_df = athletic_sales_2020_2021_womens_df.groupby(["retailer", "region", "state", "city"]).agg({"units_sold": "sum"}).rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
athletic_sales_2020_2021_womens_retailer_df.sort_values("Womens_Footwear_Units_Sold", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
athletic_sales_2020_2021_womens_pivoted_retailer_df = (pd.pivot_table(athletic_sales_2020_2021_womens_df, index=["retailer", "region", "state", "city","units_sold"], values=["total_sales"], aggfunc=sum)).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_2020_2021_womens_pivoted_retailer_df.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results.
athletic_sales_2020_2021_womens_pivoted_retailer_df.sort_values("Womens_Footwear_Units_Sold", ascending=False).head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
athletic_sales_2020_2021_day_most_sales_df = pd.pivot_table(athletic_sales_2020_2021_df, index=["invoice_date"], values=["total_sales"], aggfunc=sum)

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_2020_2021_day_most_sales_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the table.
athletic_sales_2020_2021_day_most_sales_df.sort_values("Total Sales", ascending=False).head(1)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
athletic_sales_2020_2021_day_most_sales_df.resample("D").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
(athletic_sales_2020_2021_day_most_sales_df.resample("D").sum()).sort_values("Total Sales", ascending=False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
athletic_sales_2020_2021_day_most_sales_df.resample("W").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
(athletic_sales_2020_2021_day_most_sales_df.resample("W").sum()).sort_values("Total Sales", ascending=False).head(10)
```
